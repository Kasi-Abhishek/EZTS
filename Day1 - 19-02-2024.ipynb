```python
#number reverse

n=int(input())
ans=0
if n>0:
    while n>0:
        rem=n%10
        ans=ans*10+rem
        n=n//10
    print(ans)
else:
    n=-(n)
    while n>0:
        rem=n%10
        ans=ans*10+rem
        n=n//10
    print(-(ans))
```

```python
#TV Discount using for loop

t=int(input())
for i in range(t):
    a,b,c,d=map(int,input().split())
    if (a-c) < (b-d):
        print("First")
    elif (a-c) > (b-d):
        print("Second")
    else:
        print("Any")
        
```

```python
#TV Discount using while loop

t=int(input())
while t>0:
    a,b,c,d=map(int,input().split())
    if (a-c) < (b-d):
        print("First")
    elif (a-c) > (b-d):
        print("Second")
    else:
        print("Any")
    t-=1
```

```python
#candies chef


t=int(input())
for i in range(t):
    n,x=map(int,input().split())
    if x>=n:
        print(0)
    else:
        rem=n-x
        if rem%4==0:
            print(rem//4)
        else:
            print((rem//4)+1)
    
```

```python
#Pizza Slices

t=int(input())
for i in range(t):
    p,c=map(int,input().split())
    s=c*p
    if s%4==0:
        print(s//4)
    else:
        print((s//4)+1)
    
```

```python
#Pizza Slices


t=int(input())
for i in range(t):
    n,x=map(int,input().split())
    ts=n*x
    tp=0
    while ts>4:
        tp=tp+1
        ts=ts-4
    if tp==0:
        print(tp)
    else:
        print(tp+1)
```
